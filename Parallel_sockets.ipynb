```python
host = '172.26.130.202' #Server ip
port = 8888

s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
s.bind((host, port))

print("Server Started")
```

```python
# MultiProcess here as well
def getGeocode(arrayString):
#     print(arrayString)
    if arrayString == '' or arrayString == ' ' or arrayString == '  ':
        return None
    else:
        loc = geocode(arrayString)
        if loc != None:
            return tuple(loc.point)
        else:
            return None
```

```python
def get_locName(x):
    tweet_loc_fetch = [ent.text for ent in nlp_wk(x).ents if ent.label_ in ['LOC']]
    return tweet_loc_fetch
```

```python
import pickle


while True:
    
    recieved_data, addr = s.recvfrom(1024)
    fetchedLocation = get_locName(recieved_data.decode("utf-8"))
#     recd = pickle.loads(data)
#     print(recd)
        
    print("Message from: " + str(addr))
    print("Sending: ")
    s.sendto(' '.join(fetchedLocation).encode(), addr)
    


```
